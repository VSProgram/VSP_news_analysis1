```python
import json
```

```python
with open(r'C:\Users\Alex\Desktop\VSP_news_analysis\data_scientist\data\processed\dataset_v2.3.json','r',encoding='utf-8') as file:
    data = json.load(file)
```

```python
num_unclear = 0
num_positive = 0
num_negative = 0
num_neutral = 0  
num_mixed = 0 
for i in data:
    if i['role_impact']['creditors'] == 'Unclear':
        num_unclear+=1
    if i['role_impact']['creditors'] == 'Mixed':
        num_mixed+=1
    if i['role_impact']['creditors'] == 'Neutral':
        num_neutral+=1
    if i['role_impact']['creditors'] == 'Negative':
        num_negative+=1
    if i['role_impact']['creditors'] == 'Positive':
        num_positive+=1

print(f'Unclear = {num_unclear}')
print(f'Positive = {num_positive}')
print(f'Negative = {num_negative}')
print(f'Neutral = {num_neutral}')
print(f'Mixed = {num_mixed}')

```
