```python
import json
from jsonschema import validate, ValidationError

```

```python
with open(r'C:\Users\Alex\Desktop\VSP_news_analysis\data_scientist\data\processed\schema.json','r',encoding = 'utf-8') as file:
    data_schema = json.load(file)
    
```

```python
schema = {
   "type": "object",
   "properties": {
       "name": {"type": "string"},
       "age": {"type": "number"},
       "email": {"type": "string", "format": "email"}
   },
   "required": ["name", "email"]
}
```

```python
data_schema
```

```python
with open(r'C:\Users\Alex\Desktop\VSP_news_analysis\data_scientist\data\processed\dataset_v2.1.json','r',encoding = 'utf-8') as file:

    data = json.load(file)
```

```python
num = 1
for i, item in enumerate(data):
    try:
        validate(instance=item, schema=data_schema)
    except Exception as e:
        print(f'Ошибка номер {num}')
        print(f"Ошибка в записи {i+1}: {e}")
        print('-----------------------------------------------')
        num+=1
```
